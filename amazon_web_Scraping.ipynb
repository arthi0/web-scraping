```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import requests
```

```python
user_agent='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'
```

```python
url='https://www.amazon.in/s?k=mobiles&crid=1SFLL3YQSPS8W&sprefix=mobile%2Caps%2C252&ref=nb_sb_noss_2'
```

```python
response=requests.get(url,headers={'User-Agent':user_agent,'Accept-Language':'en-US,en;q=0.5'})
```

```python
response.text
```

```python
soup=bs(response.text)
soup
```

```python
Product_name=soup.find_all('div',class_='a-section a-spacing-small a-spacing-top-small')
Product_name
```

```python
len(Product_name)
```

```python
product_name = []
for i in Product_name:
    name = i.find('h2', class_='a-size-mini a-spacing-none a-color-base s-line-clamp-2')
    if name:
        product_name.append(name.text.strip())
product_name
```

```python
rating=[]
for i in Product_name:
  rate = i.find('span',class_='a-size-base s-underline-text')
  if rate:
    rating.append(rate.text.strip())
rating
```

```python
Product_Cost=[]
for i in Product_name:
  cost= i.find('span',class_='a-offscreen')
  if cost:
    Product_Cost.append(cost.text.strip())
Product_Cost
```

```python
mrp=[]
for i in Product_name:
  m=i.find('div',class_='a-section aok-inline-block')
  if m:
    mrp.append(m.text.strip().split('₹')[1])

mrp
```

```python
dic={'Product_name':product_name,'rating':rating,'Product_Cost':Product_Cost,'Actual_cost':mrp}
```

```python
df=pd.DataFrame(dic)
```
